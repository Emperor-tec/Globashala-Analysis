```python
Campaign_spent_amount=data.groupby('Campaign Name')['Amount Spent in INR'].sum().sort_values(ascending=False)
Campaign_spent_amount=Campaign_spent_amount.reset_index()
Campaign_spent_amount
```

```python
Campaign_CPR=data.groupby('Campaign Name')['Cost per Result (CPR)'].mean().sort_values(ascending=False)
Campaign_CPR=Campaign_CPR.reset_index()
Campaign_CPR
```

```python
sns.barplot(data=Campaign_CPR,y='Campaign Name',x='Cost per Result (CPR)')
```

```python
CTR=data.groupby('Campaign Name')['Click-Through Rate (CTR)'].mean().sort_values(ascending=False)
CTR=CTR.reset_index()
CTR
```

```python
sns.barplot(data=CTR,y='Campaign Name',x='Click-Through Rate (CTR)')
```
